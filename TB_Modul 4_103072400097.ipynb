```python
import matplotlib.pyplot as plt #mengimport library

data = [
    ("Budi", 81.24, 73.66, 95.89, 2023),
    ("Siti", 98.52, 84.86, 88.7, 2024),
    ("Asep", 91.96, 71.03, 79.93, 2022),
    ("Dewi", 87.96, 97.28, 71.91, 2024),
    ("Andi", 74.68, 77.76, 79.33, 2023),
    ("Lestari", 74.68, 89.88, 79.76, 2022),
    ("Joko", 71.74, 79.35, 91.89, 2022),
    ("Rina", 95.99, 85.6, 89.13, 2022),
    ("Fajar", 88.03, 86.4, 96.62, 2024),
    ("Indah", 91.24, 75.55, 84.17, 2023),
    ("Agus", 70.62, 99.09, 73.59, 2022),
    ("Fitri", 99.1, 93.25, 91.4, 2022),
    ("Rudi", 94.97, 98.18, 92.82, 2022),
    ("Maya", 76.37, 96.84, 86.84, 2024),
    ("Hadi", 75.45, 87.94, 93.13, 2024),
    ("Sri", 75.5, 97.66, 84.81, 2023),
    ("Wawan", 79.13, 72.65, 85.68, 2024),
    ("Yuni", 85.74, 75.88, 82.83, 2022),
    ("Bambang", 82.96, 71.36, 70.76, 2023),
    ("Ayu", 78.74, 79.76, 73.24, 2022),
]

tahunMhs = [] #membuat list kosong untuk tahun per angkatan

for row in data: #looping untuk setiap baris yang ada pada data
    tahun = row[4] #menyimpan tahun (terletak pada elemen ke 5 pada array)
    if tahun not in tahunMhs: #pengecekan apakah tahun sudah ada pada list tahunMhs
        tahunMhs.append(tahun) #jika belum, di tambahkan pada list tahunMhs

tahunMhs.sort() #mengurutkan tahun

nilaiTahun = {} #membuat dictionary untuk menyimpan semua nilai dari masing" tahun
for row in data: #looping untuk setiap baris yang ada pada data
    tahun = row[4] #menyimpan tahun (terletak pada baris ke 5 pada array)
    nilai = [row[1], row[2], row[3]]  #mengambil semua nilai dari 1 mahasiswa
    if tahun not in nilaiTahun: #pengecekan apakah tahun sudah ada pada dictionary nilaiTahun
        nilaiTahun[tahun] = nilai #jika belum, membuat key baru dengan value berupa list mahasiswa
    else: 
        nilaiTahun[tahun].extend(nilai) #jika sudah, menambahkan nilai ke list tahun 

dataMhs = [nilaiTahun[t] for t in tahunMhs] #mengambil nilai berdasarkan urutan tahun, lalu disimpan dalam list dataMhs
print("=== Data Mahasiswa ===")
print()

for i in range(len(tahunMhs)): #looping akan berjalan berdasarkan jumlah tahun
    tahun = tahunMhs[i] #menyimpan tahun ke-i kedalam variabel tahun
    nilaiMhs = dataMhs[i] #mengambil semua nilai mahasiswa pada tahun ke-i

    #menghitung nilai maksimum, minimum, dan rata-rata
    nilaiMaksimum = max(nilaiMhs)
    nilaiMinimum = min(nilaiMhs)
    nilaiRataRata = sum(nilaiMhs) / len(nilaiMhs)

    #menampilkan tahun, nilai maksimum, minimum, dan rata-rata
    print("Tahun: ", tahun)
    print("Nilai Maksimum: ", nilaiMaksimum)
    print("Nilai Minimum: ", nilaiMinimum)
    print("Nilai Rata-rata: ", nilaiRataRata)
    print()

fig = plt.figure(figsize=(10,7)) #membuat kanvas dengan lebar 10 dan tinggi 7
plt.boxplot(dataMhs, tick_labels=tahunMhs, patch_artist=True) #membuat bloxplot
plt.title("Data Mahasiswa") #judul bloxplot
plt.xlabel("Tahun Angkatan") #label untuk sumbu x
plt.ylabel("Nilai") #label untuk sumbu y
plt.show() #menampilkan bloxplot
```
