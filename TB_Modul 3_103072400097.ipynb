```python
#mengimport library
import networkx as ns
import matplotlib.pyplot as plt

#fungsi showGraph
def show_graph(G, pos=None, title=''):
    if pos is None: #kalau pos kosong, posisi node akan diatur secara otomatis
        pos = ns.spring_layout(G)

    ns.draw( #menggambar graph
        G, #graph yang ingin di tampilkan
        pos, #koordinat setiap pos
        with_labels=True, #menampilkan nama node
        node_color='red', #warna node
        node_size=1500, #ukuran node
        font_size=7, #ukuran nama label
        font_color="white", #warna nama label
        font_weight="bold", #nama label dibuat tebal
        width=2 #ketebalan garis edge
    )

#edge sebagai garis penghubung antar node
    edge_labels = ns.get_edge_attributes(G, 'weight')
    ns.draw_networkx_edge_labels( #menampilkan label edge
        G, #graph yang ingin ditampilkan
        pos, #koordinat setiap pos
        edge_labels=edge_labels, #data label yang ingin ditampilkan
        font_color='blue', #warna label
        font_weight="bold", #tulisan label dibuat tebal
        font_size=7,           
    )

plt.margins(0.7)  #memberikan margin (jarak) pada plot 
```

```python
G1 = ns.Graph() #membuat graph baru

#koordinat node
pos = {
    "Arad": (-4, 2), 
    "Zerind": (-4, 3), 
    "Oradea": (-3, 4),
    "Sibiu": (-2, 2), 
    "Fagaras": (-1, 2),
    "Rimnicu Vilcea": (-2, 1), 
    "Pitesti": (-1, 1),
    "Timisoara": (-4, 1), 
    "Lugoj": (-3, 0),
    "Mehadia": (-3, -1), 
    "Drobeta": (-3, -2),
    "Craiova": (-2, -2),
    "Bucharest": (0, 0),
    "Giurgiu": (0, -1),
    "Urziceni": (1, 0),
    "Hirsova": (2, 0),
    "Eforie": (3, -1),
    "Vaslui": (2, 1),
    "Iasi": (2, 2),
    "Neamt": (1, 3)
}

#simpul
edges = [
    ("Arad", "Zerind", 75),
    ("Arad", "Timisoara", 118),
    ("Arad", "Sibiu", 140),
    ("Zerind", "Oradea", 71),
    ("Oradea", "Sibiu", 151),
    ("Timisoara", "Lugoj", 111),
    ("Lugoj", "Mehadia", 70),
    ("Mehadia", "Drobeta", 75),
    ("Drobeta", "Craiova", 120),
    ("Craiova", "Pitesti", 138),
    ("Craiova", "Rimnicu Vilcea", 146),
    ("Sibiu", "Fagaras", 99),
    ("Sibiu", "Rimnicu Vilcea", 80),
    ("Rimnicu Vilcea", "Pitesti", 97),
    ("Fagaras", "Bucharest", 211),
    ("Pitesti", "Bucharest", 101),
    ("Bucharest", "Giurgiu", 90),
    ("Bucharest", "Urziceni", 85),
    ("Urziceni", "Hirsova", 98),
    ("Hirsova", "Eforie", 86),
    ("Urziceni", "Vaslui", 142),
    ("Vaslui", "Iasi", 92),
    ("Iasi", "Neamt", 87),
]

G1.add_weighted_edges_from(edges) #menambahkan edges pada graph
show_graph(G1, pos=pos) #menampilkan graph
```

```python
shortest_paths = dict(ns.all_pairs_dijkstra_path_length(G1)) #menghitung jarak terpendek dari setiap node

for node_awal in shortest_paths: #loop dari node awal 
    for node_tujuan, jarak in shortest_paths[node_awal].items(): #loop dari node tujuan
        if node_awal != node_tujuan: #pengecekan apakah node memiliki tujuan ke dirinya sendiri
            print(f"{node_awal} -> {node_tujuan} : {jarak}") #jika tidak, menampilkan rute
    print()
```
